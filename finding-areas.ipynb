## 1.Tratamento dos dados
Seleção, limpeza e transformação dos dados

```python
# Remover acentos dos nomes dos municípios, algumas bases não vieram com acentos nos nomes, ou com nomes errados
import unidecode

def remover_acentos(texto):
    return unidecode.unidecode(texto)
```

### 1.1 Quilombolas

```python
quilombola_f = quilombola.loc[:, ["nm_municip", "nr_familia", "nr_area_ha", "geometry"]] # Selecionando algumas colunas
quilombola_f.head()
```

```python
quilombola_f.reset_index(drop=True, inplace=True) # Resetando o índice para facilitar depois
quilombola_f["nm_municip"] = quilombola_f["nm_municip"].str.lower() # Colocando o nome dos municípios em minúsculo
quilombola_f["nm_municip"] = quilombola_f["nm_municip"].apply(remover_acentos) # Removendo acentos

quilombola_f.sample(5)
```

### 1.2 CNUC

```python
cnuc_f = cnuc[cnuc.uf.str.contains("PARÁ")] # Filtrando apenas as UCs do Pará
cnuc_f.reset_index(drop=True, inplace=True) # Resetando o índice para facilitar depois
cnuc_f["municipio"] = cnuc_f["municipio"].str.lower() # Colocando o nome dos municípios em minúsculo
cnuc_f["municipio"] = cnuc_f["municipio"].apply(remover_acentos) # Removendo acentos
cnuc_f["municipio"] = cnuc_f["municipio"].str.replace("(pa)", '') # Removendo a sigla do estado para que os municípios fiquem com nomes iguais
cnuc_f = cnuc_f[["nome_uc", "ha_total", "uf", "municipio", "situacao", "geometry"]] # Selecionando algumas colunas

cnuc_f.sample(5)
```

### 1.3 SIGEF


```python
sigef_f = sigef[sigef["ind_tipo"] != "PCT"] # Filtrando para retirar CARs do tipo Território Tradicional de Povos e Comunidades Tradicionais. Estes PCTs são terras indígenas, quilombolas, etc.
sigef_f.reset_index(drop=True, inplace=True) # Resetando o índice para facilitar depois
sigef_f = sigef_f.loc[:, ["cod_imovel", "ind_tipo", "municipio", "num_area", "geometry"]] # Selecionando algumas colunas
sigef_f["municipio"] = sigef_f["municipio"].str.lower() # Colocando o nome dos municípios em minúsculo
sigef_f["municipio"] = sigef_f["municipio"].apply(remover_acentos) # Removendo acentos
sigef_f.head()
```

### 1.4 TIS

```python
tis_f = tis[tis["uf_sigla"] == "PA"] # Filtrando pelo estado do Pará
tis_f.reset_index(drop=True, inplace=True) # Resetando o índice para facilitar depois
tis_f["municipio"] = tis_f["municipio_"].str.lower() # Colocando o nome dos municípios em minúsculo
tis_f["municipio"] = tis_f["municipio"].apply(remover_acentos) # Removendo acentos
tis_f = tis_f.loc[:, ["etnia_nome", "municipio", "geometry"]] # Selecionando algumas colunas

tis_f.sample(5)
```

### 1.5 Alertas

```python
alertas_f = alertas[alertas["ESTADO"] == "PARÁ"] # Filtrando pelo estado do Pará
alertas_f.reset_index(drop=True, inplace=True) # Resetando o índice para facilitar depois
alertas_f["MUNICIPIO"] = alertas_f["MUNICIPIO"].str.lower() # Colocando o nome dos municípios em minúsculo
alertas_f["MUNICIPIO"] = alertas_f["MUNICIPIO"].apply(remover_acentos) # Removendo acentos

alertas_f.sample(5)
```

## 2. Explorando as áreas

```python
# Listar todos os municípios presentes nas bases para filtrar a base da SIGEF
def listar_mun(column, sep):  
    muns = set()
    for i in column.__iter__():
        if i.find(sep) != -1:
            i = i.split(sep)
            for j in i:
                muns.add(j)
        else:
            muns.add(i)
    return muns
```

```python
municipios_quilombolas = listar_mun(quilombola_f["nm_municip"], ' / ')
print(len(municipios_quilombolas))
```

```python
municipios_cnuc = listar_mun(cnuc_f["municipio"], ' , ')
print(len(municipios_cnuc))
```

```python
municipios_tis = listar_mun(tis_f["municipio"], ',')
print(len(municipios_tis))
```

```python
municipios = municipios_quilombolas.union(municipios_cnuc, municipios_tis) # Juntando os nomes
municipios = list(municipios)
for i in range(len(municipios)):
    municipios[i] = municipios[i].strip() # Alguns nomes estavam com um espaço no início ou no final

municipios
```

```python
sigef_f_mun = sigef_f[sigef_f["municipio"].isin(municipios)] # Filtrando base
sigef_f_mun.sample(5)
```

```python
alertas_f_mun = alertas_f[alertas_f["MUNICIPIO"].isin(municipios)] # Filtrando base
alertas_f_mun.sample(5)
```

# 3. Calculando interseções

## 3.1 Quilombola

```python
# SIGEF
areas = [] 

for i, row in quilombola_f.iterrows(): # iterando na base
    sum = 0 # Criando a variável para somar as áreas
    aux = row["nm_municip"].split(" / ") # Separando os municípios que estão em uma mesma linha, nem todas as linhas tem mais de um município
    muns = set() # Armazenando em um conjunto porque não queremos municípios repetidos
    for k in aux: # Iterando na lista, podendo ser uma lista de apenas um município
        muns.add(k.strip()) # Removendo espaços, e colocando ele no conjunto
    
    for j, row_j in sigef_f_mun[sigef_f_mun["municipio"].isin(muns)].iterrows(): # iterando na base, e filtrando apenas com os municípios que estão na linha da iteração quilombola
        sum += row["geometry"].intersection(row_j["geometry"]).area # Calculando a área de interseção entre as geometrias, e somando a área
    areas.append(sum) # Salvando a soma das áreas

areas
```

```python
# ALERTAS

areas_quilombola_alertas, count_quilombola_alertas = [], []

for i, row in quilombola_f.iterrows():
    sum_area, count_alerta = 0, 0
    aux = row["nm_municip"].split(" / ")
    muns = set()
    for k in aux:
        muns.add(k.strip())
    
    for j, row_j in alertas_f_mun[alertas_f_mun["MUNICIPIO"].isin(muns)].iterrows():
        a = row["geometry"].intersection(row_j["geometry"]).area
        if a > 0:
            count_alerta += 1
            sum_area += a
    areas_quilombola_alertas.append(sum_area)
    count_quilombola_alertas.append(count_alerta)
```

```python
# SIGEF
areas_cnuc = []

for i, row in cnuc_f.iterrows():
    sum = 0
    aux = row["municipio"].split(" , ")
    muns = set()
    for k in aux:
        muns.add(k.strip())
    
    for j, row_j in sigef_f_mun[sigef_f_mun["municipio"].isin(muns)].iterrows():
        sum += row["geometry"].intersection(row_j["geometry"]).area
    areas_cnuc.append(sum)

areas_cnuc
```

```python
# Alertas

areas_cnuc_alertas, count_cnuc_alertas = [], []

for i, row in cnuc_f.iterrows():
    sum_area, count_alerta = 0, 0
    aux = row["municipio"].split(" , ")
    muns = set()
    for k in aux:
        muns.add(k.strip())
    
    for j, row_j in alertas_f_mun[alertas_f_mun["MUNICIPIO"].isin(muns)].iterrows():
        a = row["geometry"].intersection(row_j["geometry"]).area
        if a > 0:
            count_alerta += 1
            sum_area += a
    areas_cnuc_alertas.append(sum_area)
    count_cnuc_alertas.append(count_alerta)

print(areas_cnuc_alertas)
count_cnuc_alertas
```
